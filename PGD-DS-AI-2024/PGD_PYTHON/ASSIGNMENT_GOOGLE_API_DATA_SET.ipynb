# 3- Perform pre-processing steps (Handle Missing values, Data type transformation)
  ## 3.a fill NaN and null values with N/A
  ## 3.b format the Price column as float
  ## 3.c introduce a new Currency column with USD value
  ## 3.d format the Last Updated column as ISO date string
# 4- Analyze all the columns 

```python
df['Rating'] = df['Rating'].fillna('N/A')
```

```python
df['Type'] = df['Type'].fillna('N/A')
```

```python
df['Content Rating'] = df['Content Rating'].fillna('N/A')
df['Current Ver'] = df['Current Ver'].fillna('N/A')
df['Android Ver'] = df['Android Ver'].fillna('N/A')
```

```python
df.isnull().sum()
```

```python
df[(df['Price'].str.startswith('$') == False) & (df['Price'].str.startswith('0') == False)]
```

```python
df.loc[(df['Price'].str.startswith('$') == False) & (df['Price'].str.startswith('0') == False), 'Price'] = 0
```

```python
df[(df['Price'].str.startswith('$') == False) & (df['Price'].str.startswith('0') == False)]
```

```python
df['Price'] = df['Price'].str.replace('$', '', regex=False)
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df['Currency'] = 'USD'
```

```python
df.head()
```

```python
df.insert(8, 'Currency', df.pop('Currency'))
```

```python
# Convert 'Last Updated' column to datetime
df['Last Updated'] = pd.to_datetime(df['Last Updated'], format='%B %d, %Y')
```

```python
# convert to ISO 8601
df['Last Updated'] = df['Last Updated'].dt.strftime('%Y-%m-%dT%H:%M:%S')
```
