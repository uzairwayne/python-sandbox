```python
df1['Sleep Duration'] = df1['Sleep Duration'].fillna(df1['Sleep Duration'].mode()[0])
```

```python
df1['Work/Study Hours'] = df1['Work/Study Hours'].fillna(0.0)
```

```python
df1['Financial Stress'] = df1['Financial Stress'].median()
```

```python
df1['Family History of Mental Illness'] = df1['Family History of Mental Illness'].fillna(df1['Family History of Mental Illness'].mode()[0])
```

```python
# Plot the distribution
sns.histplot(df1['Depression'], kde=False, bins=2, color='skyblue')
plt.title('Distribution of Depression Variable')
plt.xlabel('Depression')
plt.ylabel('Frequency')
plt.xticks([0, 1])  # Adjust the ticks for binary values
plt.show()
```

```python
# Example: Visualizing numerical columns in a DataFrame
for column in df1.select_dtypes(include='number').columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df1[column], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Example: Boxplot for Depression grouped by Gender
sns.boxplot(data=df1, x='Gender', y='Depression', palette='Set2')
plt.title('Boxplot of Depression by Gender')
plt.xlabel('Gender')
plt.ylabel('Depression')
plt.show()
```

```python
# Boxplot for all numerical features
for column in df1.select_dtypes(include='number').columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df1, y=column, color='lightgreen')
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.show()
```

```python
# Boxplot for Depression levels by Gender
sns.boxplot(data=df1, x='Gender', y='Depression', palette='Set2')
plt.title('Depression Levels by Gender')
plt.xlabel('Gender')
plt.ylabel('Depression')
plt.show()

# Boxplot for Depression levels by Study Habits
sns.boxplot(data=df1, x='Study_Habits', y='Depression', palette='Set3')
plt.title('Depression Levels by Study Habits')
plt.xlabel('Study Habits')
plt.ylabel('Depression')
plt.show()
```
